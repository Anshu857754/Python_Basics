```python
# IF statement

age = 20
if age >= 18:
    print("You are allowed to vote in the elections.")
```

```python
# else Statement 

age = 20
if age >= 18:
    print("You are allowed to vote in the elections.")
else:
    print("You are not allowed to vote in the elections.")
```

```python
# elif statement
age = 20
if age < 13:
    print("You are a child.")
elif age >= 18:
    print("You are teenager.")
else:
    print("You are adult.")
```

```python
# nested conditional statements
# number even odd or negative 

num = int(input("enter the number"))

if num >= 0:
    print("The number is positive")
    if num % 2 == 0:
        print("Even number")
    else:
        print("Odd number")

else:
    print("The number is negative or zero")
```

```python
# determine the year is leap year 
year = int(input("Enter the year "))

if year%4 == 0:
    if year%100 == 0:
        if year % 400 == 0:
            print(year,"is a leap year  ")
        else:
            print(year,"not a leap year")
    else:
        print(year,"is  a leap year")
else:
    print(year,"is not a leap year.")
```

```python
# Assignment 
# ismple calculator 

num1 = float(input("Enter the  firt number:-"))
num2 = float(input("Enter the  Second number:-"))
opertaions = input("enter operations('+','-','*','/','//')")

if opertaions == '+':
    result = num1+num2
elif opertaions == '-':
    result = num1-num2
elif opertaions == '*':
    result = num1*num2
elif opertaions == '/':
    if num2 !=0:
        result = num1/num2
    else:
        result = "Error! Division by zero"
else:
    result = "Invalid Operation"

print("Result:", result)
```
