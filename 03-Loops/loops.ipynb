## While Loop

```python
## The while loop continues to execute as long as the condition is True.

count = 0

while count < 5:
    print(count)
    count +=1
```

```python
count = 0 

while count % 2 == 0:
    print(count)
count +=1

"INFINITE CODE"

```

```python
# Loop Control Statement

# Break Statement

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# continue Satements

for i in range(10):
    if i %2 ==0: # iska mtlb hua ki jo bhi 2 se divide ho rhe h usko chod ker sab print kerwa do.
        continue 
    print(i)
```

```python
# Pass

## The pass statement is a null operation; it does nothing.

for i in range(5):
    if i ==3:
        print("the number is ",i)
        pass
    print(i)
```

```python
for i in range(4): # 4 class h  Class (4 classes — 0,1,2,3)
    for j in range(3): # 3 student h  Students (3 students — 0,1,2)
        print(f"i:{i} and j : {j}")



```

```python
# calcultate the sum of  first n natural numbers  by using while loop 

sum = 0
count = 1
n = int(input("Enter your number"))

while  count  <= n:
    sum = sum+count
    count +=1

print("sum of first n natural number is:",sum)
```

```python
# using for loop
sum = 0
n = int(input("Enter your number: "))

for i in range(n+1):
    sum = sum + i

print("Sum of first n natural numbers is:", sum)  # ✅ loop ke bahar
```

```python
# prime number 1 to 100

n = int(input("enter you number"))
for num in range (1,n):
    if num>1:
        for i in range (2,num):
            if num % i == 0:
                break
        else:
            print(num)
```
