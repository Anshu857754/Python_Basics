### Real-World Examples Using Lists in Python

Lists are one of the most commonly used data structures in Python, thanks to their versatility and ease of use.
Here are several practical examples that illustrate their use in real-world scenarios

## Example 1. Manage A To Do List

. Create a To Do List To Keep Track OF Tasks

```python
to_do_list = ["Buy groceries","Clean the house","Pay bills"]

## adding to task

to_do_list.append("scheduled mettings")
to_do_list.append("Go for run ")

## Removing the completed task 
to_do_list.remove("Clean the house")

## checking the task if task is in the list 

if "pay bills" in to_do_list:
    print("Don't forget to pay the utility bills")

print("To do list remmaining")
for task in to_do_list:
    print(f"-{task}")

```

### Example 2: Organizing Student Grades

        . Create a list to store and calculate average grades for students

```python
grades = [82,85,89,94,75]

# adding a new grade 
grades.append(95)

# calculating the average grade
average_grade = sum(grades)/len(grades)
print(f"Average grade {average_grade:.2f}")

# find the highest andd lowest grades
highest_grades = max(grades)
lowest_grades = min(grades)
print(highest_grades)
print(lowest_grades)
```

## Example 3: Managing An Inventory

  . Use a list to manage inventory items in a store

```python
# menninging the inventory 

inventory = ["buletooth","charger","adaptor","powerbank"]

# adding a new item in  inventory 

inventory.append("PS5")


# remove the items that is not availiable 

inventory.remove("adaptor")


#check the items are in stocks or not 
items = "adaptor"
if items in inventory:
    print(f"{items} items are in stock.")
else:
    print(f"{items} items are not in stock.")

# printing the inventory list
print("Inventory list:")

for item in inventory:
    print(f"- {item}")
```

## Example 4:Collecting User Feedback

. Use a list to collect and analyze user feedback.

```python
# Feedback 
feedback = ['Great services','amazing experince','superb experience','good quality',"Excellent service"]

# adding feedback
feedback.append("not happy with their services")

# counting the specific feedback
positive_feedback_count = sum(1 for comment in feedback
                        if "great" in comment.lower()
                        or "Excellent"in comment.lower()
)
print(f"Positive Feedback Count: {positive_feedback_count}")

# print all feedback

print("all feedback: ")

for comment in feedback:
    print(f"-{comment}")
```
