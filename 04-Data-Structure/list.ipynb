### List Comprehension

Basic syntax [expression for item in itreable] 

with conditional logic [expression for item in itreable if conditions]

Nested list Comperehension [expression for item1 in iterable1 for item2 in iterable2]
 



```python
# basic list comprehension

square = [num**2 for num in range (11)]
print(square)
```

```python
# list comprehension with conditions
lst=[]
for i in range(11):
    if i % 2 == 0:
        lst.append(i)

print(lst)
```

```python
even_numbers = [num for num in range(11) if num % 2 == 0]
print(even_numbers)
```

```python
#  Nested list Comperehension [expression for item1 in iterable1 for item2 in iterable2]

l1 = [1,2,3,4,5]
l2 = ['a','b','c','d','e']

pair = [(i,j)for i in l1 for j in l2]

print(pair)
```

```python
## List Comprehension with function calls

words = ["hello", "world", "python", "list", "comprehension"]
lengths = [len(word) for word in words]
print(lengths) # Output: [5,5,6,4,13]
```

#### Conclusion

List comprehensions are a powerful and concise way to create lists in Python. They are syntactically compact
and can replace more verbose looping constructs. Understanding the syntax of list comprehensions will help
