## Lambda Functions in Python

Lambda functions are small anonymous functions defined using the lambda keyword. They can have any
number of arguments but only one expression. They are commonly used for short operations or as arguments
to higher-order functions.

```python
addition = lambda a,b:a+b 
type(addition)
print(addition(5,9))
```

```python
def even (num):
    if num % 2 == 0:
        return True
    

even(24)
```

```python
even1 = lambda num:num%2 == 0
print(even1(67))
```

```python
def addition(x,y,z):
    return x+y+z

addition(3,6,7)
```

```python
addi = lambda x,y,z:x+y+z
print(addi(2,4,6))
```

```python
## Map ()

numbers = [1,2,3,4,5,6,7]
def squares (numbers):
    return numbers**2

squares(12)
```

```python
map(lambda x:x**2, numbers)
# map commonly used for lazy loading technique
```

```python
squa = lambda x:x**2
squa(5)
```
