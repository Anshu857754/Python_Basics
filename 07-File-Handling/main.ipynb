# File Operation- Read And Write Files

File handling is a crucial part of any programming language. Python provides built-in functions and methods to
read from and write to files, both text and binary. This lesson will cover the basics of file handling, including
reading and writing text files and binary files.

```python
# read whole file 

with open('example.txt','r')as file: # r ka use hum read ke liye kerte h 
    content = file.read()
    print(content)
```

```python
# read file line by line
with open('example.txt','r')as file:
    for line in file:
        print(line.strip()) # .strip new line character ko hata deta h.

```

```python
# Writting a file overwritting
with open('example.txt','w')as file:
    file.write("Hello kya hal h ?\n")
    file.write('this is a new line\n')
    # agr m isko run kerunga to purane jitnna bhi content hoga usko overwritten ker dega.
```

```python
# without overwitting
with open('example.txt','a')as file: # a ka use hum append ke liye kerte h .
    file.write("without overwritting append operation take place\n")
```

```python
# writting a list of line in a files 
line = ['first-line \n','second-line \n','third-line']
with open('example.txt','a')as file:
    file.writelines(line)
```

```python
# creating a  bin file
### Binary Files

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file: # wb ka mtlb h write bite mode or rb means read bite mode
    file.write(data)
```

```python
# read the context  from the source text file and write to a destination text file 
# copying a text file

with open('example.txt','r') as source_file:
    content = source_file.read()

with open('destination.txt','w')as destination_file:
    destination_file.write(content)
```

```python
def count_text_file(file_path):
    with open(file_path, 'r') as file:          
        lines = file.readlines()                
        line_count = len(lines)                 
        word_count = sum(len(line.split()) for line in lines)  
        char_count = sum(len(line) for line in lines)          
        return lines, line_count, word_count, char_count

file_path = 'example.txt'
lines, line_count, words, characters = count_text_file(file_path) 
print(f"Lines: {line_count}, Words: {words}, Characters: {characters}")
```

```python
# writting and then readings the files 

with open ('example.txt','w+') as file:
    file.write("hello world \n")
    file.write("this is a new line \n")

    file.seek(0) # move the file coursor to the begning 

    # read the content of the file
    content=file.read()
    print(content)
```
