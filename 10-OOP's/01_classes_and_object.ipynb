## Classes and Objects

Object-Oriented Programming (OOP) is a programming paradigm that uses "objects" to design applications
and computer programs. OOP allows for modeling real-world scenarios using classes and objects. This lesson
covers the basics of creating classes and objects, including instance variables and methods.

```python
# A classes is a blue print for creating a objects. Attributes and methods.

# create a class

class car :
    pass
audi = car()
bmw = car() # audi is a objects

print(type(audi))
```

```python
# instance method or instance variable

class Dog:
    #  constructer
    def __init__(self,name,age): # give the two attributes 
            self.name = name
            self.age =  age

#  create objects 

dog1 = Dog("Toto",12)

print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
dog2 = Dog("Luci",5)
print(dog2.name)
print(dog2.age)
```

```python
# Define a class with a instance method
class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def bark(self):
        print(f"{self.name} say Woof")

dog1 = Dog("Toto",12)
dog1.bark()
dog2 = Dog("Luci",5)
dog2.bark()
```

```python
# bank Amount 

class BankAccount:
    def __init__(Bank,Owner,Balance = 0):
        Bank.Owner = Owner
        Bank.Balance = Balance
    def deposite(Bank,Amount):
        Bank.Amount = Amount
        Bank.Balance+=Amount
        print(f"{Amount} is deposited.New balance is {Bank.Balance}")
    def withdraw(Bank,Amount):
        if Amount>Bank.Balance:
            print("Ifficiant Funds")
        else:
            Bank.Balance-=Amount
            print(f"{Amount} is Withdraw. Current balance is {Bank.Balance}")
    def get_balance(Bank):
        return Bank.Balance
    
# create an account 

account = BankAccount("Anshu",50000)
print(account.Balance)


```
