## Inheritance In Python

Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class to inherit
attributes and methods from another class. This lesson covers single inheritance and multiple inheritance,
demonstrating how to create and use them in Pvthon.

```python
# inheritance

## Parent class  

class Car:
    def __init__(self,windows,doors,engine_type):
        self.windows = windows 
        self.doors = doors
        self.engine_type = engine_type

    def drive(self):
        print(f"the person will drive the {self.engine_type} car")

```

```python
car1 = Car(4,5,"Petrol")
print(car1.drive())
```

```python
class Tesla(Car):
    def __init__(self,windows,doors,engine_type,is_selfdriving):
        super().__init__(windows,doors,engine_type)
        self.is_selfdriving = is_selfdriving

    
    def selfdriving(self):
        print(f"tesla Supports self driving : {self.is_selfdriving}")
```

```python
tesla1 = Tesla(4,5,"electric",True)
print(tesla1.selfdriving())
```

```python
tesla1.drive()
```

```python
# IT means that one class is inherit from diffrent classes 

# Multiple Inheritance
## When a class inherits from more than one base class.
# base1
class Animal():
    def __init__(self,name):
        self.name = name
    
    def speak(self):
        print("Subclass must implement this python")

# base2 ;
class Pet:
    def __init__(self,owner):
        self.owner = owner

# Derived Class
class Dog(Animal,Pet):
    def __init__(self, name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    
    def speak(self):
        return f"{self.name} say woof"
    

# create an object 
dog  = Dog("Buddy","Anshu")
print(dog.speak())
print(f"Owner:{dog.owner}")
```
