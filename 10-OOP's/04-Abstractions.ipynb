## Abstraction

Abstraction is the concept of hiding the complex implementation details and showing only the necessary
features of an object. This helps in reducing programming complexity and effort.

# # Working flow

Vechile (Abstract)

    ↓ inherit

   Car

    ↓ object

   car

    ↓ pass

Operate_vechile(car)

    ↓

start_engine() → "The engine is started"

drive()        → "The vehicle is used for driving"

```python
# Abstract class banane ke liye import karna zaruri ha ye humari python ki inbuilt class m aata h isko hum vhnage nhi ker sakte h.
from abc import ABC,abstractmethod  


# abstract class

class Vechile(ABC): # ABC se inherit kiya — ab yeh abstract hai
    def drive(self):
        print("The vichle is used for driving") # Yeh normal method hai — override karna zaruri nahi 
    
    # @abstractmethod matlab → "Beta! Yeh method apne aap likhna hoga!"
    @abstractmethod 
# Yeh blueprint hai — Child class mein override karna ZARURI hai
    def start_engine(self):
        pass

# child Class 
class Car(Vechile):
    def start_engine(self):
        print("the engine is started")
# Car ne start_engine implement kiya — ab object ban sakta hai

# Koi bhi Vechile object pass karo — dono kaam karega!
def Operate_vechile(vechile):
    vechile.start_engine()
    vechile.drive()

car = Car()
Operate_vechile(car)
```
