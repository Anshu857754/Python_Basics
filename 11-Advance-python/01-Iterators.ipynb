## Iterators

Iterators are advanced Python concepts that allow for efficient looping and memory management. Iterators
provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
my_list = [1,2,3,4,5,6]
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
# create the iterator 
iterator = iter(my_list)
print(type(iterator))
```

```python
# by using a lazy loading techniqe

# iterate through all the element 
next(iterator)
```

```python
iterator = iter(my_list)
```

```python
# handel the lazy loading techniqe by using the exception handeling

try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterators")
```
