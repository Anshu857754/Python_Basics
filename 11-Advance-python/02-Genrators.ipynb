## Generators

Generators are a simpler way to create iterators. They use the yield keyword to produce a series of values lazily,
which means they generate values on the fly and do not store them in memory.

```python
# it is the subclass of the iterator

def square(n):
    for i in range(3):
        yield i ** 2
        # yield keyword is also returning the values
```

```python
square(2)
```

```python
for i in square(3):
    print(i)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3

```

```python
gen = my_generator()
gen
```

```python
for val in gen:
    print(val)
```
