## Pandas-DataFrame And Series

Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It
provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object,
while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure
with labeled axes (rows and columns).

```python
# series 
# it is a one dimensional data type that can create easily 
import pandas as pd
data = [1,2,3,4,5,6,7,8,9]
series = pd.Series(data)
print("series",series)
```

```python
# create the series from dictionary 
import pandas as pd
data={'a':1,'b':2,'c':3,'d':4,'e':5}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [1,2,3,4]
index = ['a','b','c','d']
pd.Series(data,index=index)
```

## DataFrames

```python
# Create the dataframe from  a dictionary of list 

data={
'Name' : ['Krish', ' John', ' Jack' ],
'Age': [25,30,45],
'City': ['Bangalore', 'New York', 'Florida']

}
df = pd.DataFrame(data)
print(df)
```

```python
# Create the data frames using  a list of Dictonary 

data = [
        {'Name': 'Krish', 'Age': 32, 'City': 'Bangalore'},
        {'Name': 'Ansh', 'Age': 19, 'City': 'delhi'},
        {'Name' : 'Anuj', 'Age' :20,'City':'pune'},
        {'Name': 'Abhay', 'Age': 21, 'City': 'Bangalore'},
        {'Name': 'Ashish', 'Age': 23, 'City': 'jaipur'},


]
df = pd.DataFrame(data)
print(df)
```
