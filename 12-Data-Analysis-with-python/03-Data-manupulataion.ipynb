```python
# Date Aggrigation and Groupping 

df.head()
```

```python
grouped_mean=df.groupby('Product') ['Value' ].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Product', 'Region' ]) ['Value' ] . sum()
print(grouped_sum)
```

```python
# agregate multiple functions 

grouped_agg=df.groupby('Region')['Value' ].agg(['mean','sum','count'])
grouped_agg
```

```python
# merging and joining DataFrame

df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2 = pd.DataFrame({'Key':['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
df1
```

```python
df2
```

```python
# merege the Dataframe on the key column 
pd.merge(df1,df2,on="Key",how="inner")
```

```python
# on="ID" — yeh batata hai ki kaunse column se dono tables ko jodna hai. Dono tables mein ID column common tha, toh on="ID" likha.
# how="inner" — yeh batata hai kaise jodna hai:

# inner → sirf common rows
# left → df1 ki sab rows + match wali df2 ki rows
# right → df2 ki sab rows + match wali df1 ki rows
# outer → dono ki sab rows, missing mein NaN
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left")
```

```python
pd.merge(df1,df2,on="Key",how="right")
```
