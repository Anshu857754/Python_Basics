## Data Visualization With Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing
attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few
lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and
customizing them.

```python
import seaborn as sns
import pandas as pd
tips = pd.read_csv('tips_dataset.csv', index_col=0)
```

```python
# basic plotting seaborn 
tips=sns.load_dataset('tips')
tips
```

```python
# create a sccater  plot
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips') 
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# line plot 
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("line plot of total bill by size")
plt.show()
```

```python
# categorical plots 
# categorical variables wo hota h jissai ki day , hogya ,smoker, time,gender(sex) 
# Jis column mein limited, fixed categories hon — woh categorical variable hota hai
#bar plots 
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x='day',y='total_bill',data=tips)
plt.title("line plot of total bill by Day")
plt.show()
```

```python
# box plots
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("line plot of total bill by Day")
plt.show()
```

```python
# Voilon plot 
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("line plot of total bill by Day")
plt.show()
```

```python
# histogram 
sns.histplot(tips['total_bill'], bins=10, kde=True)
plt.show()
bins=10

# bins = 10
# Poore data ko 10 equal hisson mein baant do
# Jaise 0-5, 5-10, 10-15... har hissa ek bar hai.


# KDE = Kernel Density Estimate
# Histogram ke upar ek smooth curve draw karta hai
# Ye dikhata hai data ka overall shape/pattern.
```

```python
# kde plot 
sns.kdeplot(tips['total_bill'],shade=True)

# kde ka use hum curve dekhne ke liye kerte h.

```

```python
# pairlot
sns.pairplot(tips)
# 💡 Sochlo: Agar 5 columns hain toh manually 10 scatter plots banane padte — pairplot ek second mein sab dikhata hai.
```

```python
# heatmap
corr = tips[['total_bill','tip','size']].corr() # corelation create kerta h 
corr
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
