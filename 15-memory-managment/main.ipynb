## Reference Counting

Reference counting is the primary method Python uses to managt memory. Each object in Python maintains a
count of references pointing to it. When the reference count drops to zero, the memory occupied by the object
is deallocated.

```python
## Refrence Counting 
# Reference Counting = Python track karta hai ki ek object ko kitni jagah se point kiya ja raha hai
import sys

a = []
print(sys.getrefcount(a))
```

```python
b = a 
print(sys.getrefcount(a))
```

```python
c=a
print(sys.getrefcount(a))
```

```python
del c 
print(sys.getrefcount(a))
```

```python
del b 
print(sys.getrefcount(a))
```

## Garbage Collection
"Garbage Collection = Python ka automatic kaam waala banda — jo bekaar objects dhundh ke memory saaf karta hai, tum tension mat lo!" 🚀

```python
# Python includes a cyclic garbage collector to handle reference cycles.
#  Reference cycles occur when objects refrence each other , preventing their refrence count from reaching zero.

import gc 
# enables the garbage collecctor 
gc.enable()
```

```python
gc.disable()
```

```python
gc.collect()
# 739 this number are unreachable 
```

```python
# get garbage collection stats
print(gc.get_stats())
```

```python
# get the unreachable objects are 
print(gc.garbage)
# now it is  a completely empty.
```

## Memory Management Best Practices

1. Use Local Variables: Local variables have a shorter lifespan and are freed sooner than global variables.

2. Avoid Circular References: Circular references can lead to memory leaks if not properly managed.

3. Use Generators: Generators produce items one at a time and only keep one item in memory at a time,
making them memory efficient.

4. Explicitly Delete Objects: Use the del statement to delete variables and objects explicitly.

5. Profile Memory Usage: Use memory profiling tools like tracemalloc and memory_profiler to identify
memory leaks and optimize memory usage.

```python
# handling a circular refrence 
import gc 

class Myobject:
    def __init__(self,name):
        self.name = name
        print(f"Object {self.name} created")

    def __del__(self):
        print(f"Object {self.name} deleted")

# create a circular refrence 
obj1 = Myobject("obj1")
obj2 = Myobject("obj2")
# isi ko hum circular refrence kehte h.
obj1.ref = obj2
obj2.ref = obj1


del obj1
del obj2

# manually trigger the garbage collection 
gc.collect()
```

```python
## Generators For Memeory Efficiency
#Generators allow you to produce items one at a time, using memory efficiently by only keeping one i

def generate_numbers(n):
    for i in range(n):
        yield i

## using the generator
for num in generate_numbers(100000):
    print(num)
    if num>10:
        break
```
